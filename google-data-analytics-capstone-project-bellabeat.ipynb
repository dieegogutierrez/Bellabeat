#### <span style="color: #f98871;">3.4 Data Limitations & Integrity</span><a class="anchor" id="datalimitationsandintegrity"></a>
* **Temporal Limitations:**

The dataset was collected in 2016, which means it may not reflect the most current fitness habits and trends. Fitness habits and technology have evolved over time, and what was popular in 2016 may not hold true today. Therefore, any analysis based on this dataset should be viewed with the understanding that it may not fully represent the current fitness landscape.

* **Sample Size:**

Another important factor to consider is the sample size. The dataset is based on the fitness data of approximately 30 users. While it offers a snapshot of the fitness habits of these individuals, it is relatively small in terms of data analysis. The Central Limit Theorem (CLT) tells us that larger sample sizes are often preferable for robust statistical analysis. With only 30 users, the dataset is close to the minimum recommended for comprehensive analysis, and findings should be interpreted with this limitation in mind.

* **Data Quality:**

Data quality is always a concern when working with datasets. Incomplete, inaccurate, or missing data can impact the validity of any analysis. Therefore, the empty values indentified before will be addressed in the next step.

* **Bias:**

The dataset includes data from a specific group of users. It is essential to acknowledge that the findings may not be generalizable to a broader population. The sample may exhibit certain biases or characteristics that are not representative of the entire population. Any generalizations drawn from this dataset should be made with caution.

#### <span style="color: #f98871;">4.2 Data Cleaning</span><a class="anchor" id="datacleaning"></a>
It is evident from the data summary that all tables share a common 'Id' column and a similar date column. These columns will be used to join the tables, reducing the total from 5 to 2 tables, one for daily data **dailyActivity** and another for hourly data **hourlyActivity**. Moreover, columns for days (i.e., Monday, Tuesday, etc.) and periods (i.e., Morning, Afternoon, etc.) will be created.

```sql
CREATE OR REPLACE TABLE `steadfast-slate-398812.Bellabeat.dailyActivity` AS
SELECT 
  day.Id,
  day.ActivityDate,
  FORMAT_DATE('%A', day.ActivityDate) AS ActivityDay,
  day.SedentaryMinutes,
  day.LightlyActiveMinutes,
  day.FairlyActiveMinutes,
  day.VeryActiveMinutes,
  day.SedentaryActiveDistance,
  day.LightActiveDistance,
  day.ModeratelyActiveDistance,
  day.VeryActiveDistance,
  day.LoggedActivitiesDistance,
  day.TrackerDistance,
  day.TotalDistance,
  day.TotalSteps,
  day.Calories,
  sleep.TotalSleepRecords,
  sleep.TotalMinutesAsleep,
  sleep.TotalTimeInBed
FROM `steadfast-slate-398812.Bellabeat.dailyActivity` AS day
LEFT JOIN `steadfast-slate-398812.Bellabeat.sleepDay` AS sleep
ON day.Id = sleep.Id AND day.ActivityDate = sleep.SleepDay
```
* Create new hourlyActivity table from the other 3 tables and split ActivityHour column into Date and Time columns.
* Create a period columns with day periods, i.e., Morning, Afternoon, etc.

```sql
CREATE OR REPLACE TABLE `steadfast-slate-398812.Bellabeat.hourlyActivity` AS
SELECT 
  cal.Id,
  DATE(cal.ActivityHour) AS ActivityDate,
  TIME(cal.ActivityHour) AS ActivityHour,
  FORMAT_DATE('%A', cal.ActivityHour) AS ActivityDay,
  CASE
    WHEN EXTRACT(HOUR FROM cal.ActivityHour) BETWEEN 5 AND 11 THEN 'Morning'
    WHEN EXTRACT(HOUR FROM cal.ActivityHour) BETWEEN 12 AND 18 THEN 'Afternoon'
    WHEN EXTRACT(HOUR FROM cal.ActivityHour) BETWEEN 19 AND 22 THEN 'Evening'
    ELSE 'Night'
  END AS Period,
  cal.Calories,
  intens.TotalIntensity,
  intens.AverageIntensity,
  step.StepTotal
FROM `steadfast-slate-398812.Bellabeat.hourlyCalories` AS cal
FULL OUTER JOIN `steadfast-slate-398812.Bellabeat.hourlyIntensities` AS intens
ON cal.Id = intens.Id AND cal.ActivityHour = intens.ActivityHour
FULL OUTER JOIN `steadfast-slate-398812.Bellabeat.hourlySteps` AS step
ON cal.Id = step.Id AND cal.ActivityHour = step.ActivityHour
```

* **Data Cleaning & Manipulation Process:**

    * **Deletion:**
    
        * Duplicates rows in the sleepDay table.
        
    * **Addition:**
        
        * ActivityDay, ActivityDate, ActivityHour and Period:

            * ActivityDay: Calculated by extracting the weekday from columns 'ActivityDate' or 'ActivityHour' (e.g., “Monday”).
            * ActivityDate: Calculated by extracting the date from 'ActivityHour' datetime column.
            * ActivityHour: Calculated by extracting the time from 'ActivityHour' datetime column.
            * Period: This column is determined based on the time of day in the 'ActivityHour' column. 
                * Morning: 5AM to 11AM
                * Afternoon: 12PM to 6PM
                * Evening: 7PM to 10PM
                * Night: 11PM to 4AM
                
    * **Creation:**
    
        * New dailyActivity table with sleepDay data added to its users.
        * Created hourlyActivity table with data from hourlyCalories, hourlyIntensities and hourlySteps.

The next step involves examining the minimum, maximum, and average values of numeric columns to gain insights into the data and identify any potential outliers that might affect the analysis. The full result of the query can be seen in [Google Sheets](https://docs.google.com/spreadsheets/d/1FTFGq8OhiXp-VyQv57mOS95JT5tLp-fzyPgc6zHu1qg/edit?usp=sharing).

```sql
SELECT 
  MIN(SedentaryMinutes) AS MIN_SedentaryMinutes,
  MAX(SedentaryMinutes) AS MAX_SedentaryMinutes,
  AVG(SedentaryMinutes) AS AVG_SedentaryMinutes,
  MIN(LightlyActiveMinutes) AS MIN_LightlyActiveMinutes,
  MAX(LightlyActiveMinutes) AS MAX_LightlyActiveMinutes,
  AVG(LightlyActiveMinutes) AS AVG_LightlyActiveMinutes,
  MIN(FairlyActiveMinutes) AS MIN_FairlyActiveMinutes,
  MAX(FairlyActiveMinutes) AS MAX_FairlyActiveMinutes,
  AVG(FairlyActiveMinutes) AS AVG_FairlyActiveMinutes,
  MIN(VeryActiveMinutes) AS MIN_VeryActiveMinutes,
  MAX(VeryActiveMinutes) AS MAX_VeryActiveMinutes,
  AVG(VeryActiveMinutes) AS AVG_VeryActiveMinutes,
  MIN(SedentaryActiveDistance) AS MIN_SedentaryActiveDistance,
  MAX(SedentaryActiveDistance) AS MAX_SedentaryActiveDistance,
  AVG(SedentaryActiveDistance) AS AVG_SedentaryActiveDistance,
  MIN(LightActiveDistance) AS MIN_LightActiveDistance,
  MAX(LightActiveDistance) AS MAX_LightActiveDistance,
  AVG(LightActiveDistance) AS AVG_LightActiveDistance,
  MIN(ModeratelyActiveDistance) AS MIN_ModeratelyActiveDistance,
  MAX(ModeratelyActiveDistance) AS MAX_ModeratelyActiveDistance,
  AVG(ModeratelyActiveDistance) AS AVG_ModeratelyActiveDistance,
  MIN(VeryActiveDistance) AS MIN_VeryActiveDistance,
  MAX(VeryActiveDistance) AS MAX_VeryActiveDistance,
  AVG(VeryActiveDistance) AS AVG_VeryActiveDistance,
  MIN(TrackerDistance) AS MIN_TrackerDistance,
  MAX(TrackerDistance) AS MAX_TrackerDistance,
  AVG(TrackerDistance) AS AVG_TrackerDistance,
  MIN(TotalDistance) AS MIN_TotalDistance,
  MAX(TotalDistance) AS MAX_TotalDistance,
  AVG(TotalDistance) AS AVG_TotalDistance,
  MIN(TotalSteps) AS MIN_TotalSteps,
  MAX(TotalSteps) AS MAX_TotalSteps,
  AVG(TotalSteps) AS AVG_TotalSteps, 
  MIN(Calories) AS MIN_Calories,
  MAX(Calories) AS MAX_Calories,
  AVG(Calories) AS AVG_Calories,
  MIN(TotalSleepRecords) AS MIN_TotalSleepRecords,
  MAX(TotalSleepRecords) AS MAX_TotalSleepRecords,
  AVG(TotalSleepRecords) AS AVG_TotalSleepRecords,
  MIN(TotalMinutesAsleep) AS MIN_TotalMinutesAsleep,
  MAX(TotalMinutesAsleep) AS MAX_TotalMinutesAsleep,
  AVG(TotalMinutesAsleep) AS AVG_TotalMinutesAsleep,
  MIN(TotalTimeInBed) AS MIN_TotalTimeInBed,
  MAX(TotalTimeInBed) AS MAX_TotalTimeInBed,
  AVG(TotalTimeInBed) AS AVG_TotalTimeInBed
FROM `steadfast-slate-398812.Bellabeat.dailyActivity`
  ```
The average sedentary time is 990 minutes, which is equivalent to 16.5 hours. In comparison, the average sleep duration is 420 minutes, or 7 hours. This indicates that users have 9.5 hours of sedentary time, which could be improved. Other important data is that, on average, users take around 7.5k steps and burn the equivalent of 2.3k calories per day.

No extreme outliers were found in the last step; therefore, the average of activity minutes will be used to check if users are reaching the WHO guidelines. The columns 'AVG_FairlyActiveMinutes' and 'AVG_VeryActiveMinutes' will be summed, and the result will be displayed in a new column called 'AVG_TotalActiveMinutes'. The full result of the query can be seen in [Google Sheets](https://docs.google.com/spreadsheets/d/1tEXx7lPXpbb0ERiUpUH2pkec0dHBygM8GxabjIFjRvk/edit?usp=sharing).

* WHO guideline - Combination of moderate and vigorous physical activity with at least 150 min/week or around 20 min/day.

```sql
SELECT
  Id,
  ROUND(AVG(FairlyActiveMinutes) + AVG(VeryActiveMinutes)) AS AVG_TotalActiveMinutes,
  CASE
    WHEN (AVG(FairlyActiveMinutes) + AVG(VeryActiveMinutes)) >= 20 THEN 'APPROVED'
    ELSE 'NOT APPROVED'
  END AS WHO_Recommendation
FROM `steadfast-slate-398812.Bellabeat.dailyActivity`
GROUP BY Id
```

On average, 13 users did not reach the minimum 20 minutes of moderate/vigorous physical activities WHO guideline. The next analysis is about the average steps, calories, sleep and total distance that each user has per day. Additionally, a column with the percentage of the distance that was very active will be added. This column will be calculated by dividing the 'VeryActiveDistance' by the total distance. The full result of the query can be seen in [Google Sheets](https://docs.google.com/spreadsheets/d/1KsXdlE0HwenxNO9_edzduoAs4e4VdiPMSdSL091BN2o/edit?usp=sharing).

```sql
SELECT 
  Id,
  ROUND(AVG(TotalSteps)) AS AVG_TotalSteps,
  ROUND(AVG(Calories)) AS AVG_Calories,
  ROUND(AVG(TotalDistance)) AS AVG_TotalDistance,
  ROUND((AVG(VeryActiveDistance) / AVG(TotalDistance)) * 100) AS PercentVeryActiveDistance,
  ROUND(AVG(TotalMinutesAsleep)) AS AVG_TotalMinutesAsleep
FROM `steadfast-slate-398812.Bellabeat.dailyActivity`
GROUP BY Id
```

Among the users, 14 individuals can be considered active as they averaged over 8,000 steps per day, with one user standing out with an impressive average of over 16,000 daily steps.
On the other hand, 5 users did not manage to reach 3,000 steps per day, indicating a less active lifestyle.

Notably, it appears that there is a relationship between calorie expenditure and the number of steps taken. Only users who achieved over 8,000 steps per day managed to burn more than 3,000 calories.

Users who took more steps generally tend to engage in more active activities. At least 20% of their distance was classified as "very active".

It's worth noting that some users reported an average of only 60-70 minutes of daily sleep, which may seem unusual.
However, when looking at the median value in the column stats, it becomes clear that the majority of users slept around 420 minutes, which is 7 hours. This aligns with the CDC guidelines for recommended sleep duration.

The final analysis will utilize the **hourlyActivity** data to determine on which weekdays, periods of day, and at what hours users are most active.

* **Weekday**: The full result of the query can be seen in [Google Sheets](https://docs.google.com/spreadsheets/d/1vHa5xJPr5i1EiryQcE5eKTOPhLwvPF4CykwGoMfhigA/edit?usp=sharing).

```sql
SELECT 
  ActivityDay,
  ROUND(AVG(Calories)) AS AVG_Calories,
  ROUND(AVG(TotalIntensity)) AS AVG_TotalIntensity,
  ROUND(AVG(StepTotal)) AS AVG_StepTotal
FROM (
  SELECT 
    Id,
    ActivityDay,
    ActivityDate,
    SUM(Calories) AS Calories,
    SUM(TotalIntensity) AS TotalIntensity,
    SUM(StepTotal) AS StepTotal
  FROM `steadfast-slate-398812.Bellabeat.hourlyActivity` 
  GROUP BY Id, ActivityDay, ActivityDate
) 
GROUP BY ActivityDay
ORDER BY
  CASE
    WHEN ActivityDay = 'Monday' THEN 1
    WHEN ActivityDay = 'Tuesday' THEN 2
    WHEN ActivityDay = 'Wednesday' THEN 3
    WHEN ActivityDay = 'Thursday' THEN 4
    WHEN ActivityDay = 'Friday' THEN 5
    WHEN ActivityDay = 'Saturday' THEN 6
    WHEN ActivityDay = 'Sunday' THEN 7
  END
```

* **Period**: The full result of the query can be seen in [Google Sheets](https://docs.google.com/spreadsheets/d/1O_Xb6J5VAUDp_QIXY8BqTR1F_cnQOAJa4KaRAC8JjAA/edit?usp=sharing).

```sql
SELECT 
  Period,
  ROUND(AVG(Calories)) AS AVG_Calories,
  ROUND(AVG(TotalIntensity)) AS AVG_TotalIntensity,
  ROUND(AVG(StepTotal)) AS AVG_StepTotal
FROM (
  SELECT 
    Id,
    Period,
    ActivityDate,
    SUM(Calories) AS Calories,
    SUM(TotalIntensity) AS TotalIntensity,
    SUM(StepTotal) AS StepTotal
  FROM `steadfast-slate-398812.Bellabeat.hourlyActivity` 
  GROUP BY Id, Period, ActivityDate
) 
GROUP BY Period
ORDER BY
  CASE
    WHEN Period = 'Morning' THEN 1
    WHEN Period = 'Afternoon' THEN 2
    WHEN Period = 'Evening' THEN 3
    When Period = 'Night' THEN 4
  END
```

* **Hour**: The full result of the query can be seen in [Google Sheets](https://docs.google.com/spreadsheets/d/1wp_RbnDKh4ciZrXmE2Teq6FOxI7NANSZ7oxaTkpwRnk/edit?usp=sharing).
